```python
## BeautifulSoup 

from bs4 import BeautifulSoup
```

```python
url = "https://wuzzuf.net/a/Website-Jobs-in-Egypt"
page = requests.get(url).text
soup = BeautifulSoup(page, "html.parser")
```

```python
print(soup)
```

```python
find = soup.find("h2")
print(find.text)
```

```python
## prettify organize outputshape
print(find.prettify())
```

```python
find = soup.find_all("div")
print(find)
```

```python
find = soup.find_all("a")
for link in find:
    print(link)
```

```python
## by Id

biId = soup.find(id="app")
print(biId.prettify())
```

```python
byId2 = soup.find("h3", id="app")
print(byId2)
```

```python
allId = soup.find_all("h2", id="app")
print(allId)
```

```python
byClass= soup.find("div", class_="css-abmyix")
print(byClass)
```
